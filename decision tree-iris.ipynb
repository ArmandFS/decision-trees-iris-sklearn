## Decision Treees in using SKLearn and the Iris Dataset
#### This short repo will be displaying how Decision Trees work in Machine Learning, I will solely use the SKLearn (Sci-Kit Learn library) to commence and run this decision tree
#### We will be using the iris dataset (which is a sort of flower taxonomy dataset)
## So what are Decision Trees?
#### Decision Trees or DTs are a non parametric supervised learning method used mostly for classification and regression. The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features
#### The sklearn library will provide the iris dataset, and will provide the sklearn function. Thee model selection will also add a function that will split the data into training and testing datasets


```python
#Import sklearn libraries
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
# Visualize the decision tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

## Loading the dataset and Training the Model
#### Here we will be getting our iris dataset and proceed to get our testing and training inputs and targets
#### train_test_split function will split the data.
#### We will initialize the DecisionTreeClassifier (a class) and make it into a variable that is easily codeable


```python
#load the iris dataset
iris = load_iris()
#split iris data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)
# Initialize the decision tree classifier
dtc = DecisionTreeClassifier()
```

## Fitting the model, running the model, and predicting accuracy
#### We can use the dtc.fit function to train the model, and after training we can predict our target label based on X_test
#### the score function will print out our model's accuracy

```python
#fit the model on the training data
dtc.fit(X_train, y_train)
tree_depth = dtc.get_depth()
print("Depth of the decision tree:", tree_depth)
#predict on the testing data
y_pred = dtc.predict(X_test)
#Evaluate the model's accuracy by using the score function
accuracy = dtc.score(X_test, y_test)
print(f"Accuracy: {accuracy}")
```

## Running and plotting the model
#### After predicting and printing our our model's accuracy, we can plot the model using matplotlib and then show the model

```python

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi=300)
plot_tree(dtc)
plt.show()
```

#### The tree graph that was outputted by the last cell of code is a visual representation of the decision tree that was learned by the model. The tree graph is constructed such that the root node is at the top of the tree, and the leaf nodes are at the bottom of the tree.

#### Each node in the tree graph represents a decision rule based on one of the input features. The decision rule is represented as a threshold value, which is shown in the box at the top of the node. For example, in the root node of the tree, the decision rule is "petal length (cm) <= 2.45".

#### The edges in the tree graph represent the decision outcomes. If the decision rule is true, then the model follows the edge labeled "True" to the next node in the tree. If the decision rule is false, then the model follows the edge labeled "False" to the next node in the tree.
